```python
## 2-snake ##
n = 2

# dynamic variables
q0, q1 = me.dynamicsymbols('q0 q1')
u0, u1 = me.dynamicsymbols('u0 u1')

q3, q4 = me.dynamicsymbols('q3 q4')
u3, u4 = me.dynamicsymbols('u3 u4')

f1 = me.dynamicsymbols('f1')
f1d = me.dynamicsymbols('f1', 1)
f1dd = me.dynamicsymbols('f1', 2)

# variables for bodies
g, t = sym.symbols('g t')
m0, m1 = sym.symbols('m_0 m_1')
l0, l1 = sym.symbols('l_0 l_1')
Izz0, Izz1 = sym.symbols('I_zz_0 I_zz_1')

# inertial reference frame
frN = me.ReferenceFrame('fr_N')

# body-fixed frames
fr0 = frN.orientnew('fr_0', 'Axis', [q0, frN.z])
fr1 = fr0.orientnew('fr_1', 'Axis', [q1, fr0.z])
#fr2 = fr1.orientnew('fr_2', 'Axis', [q2, fr1.z])

# angular velocities of frames
fr0.set_ang_vel(frN, u0 * frN.z)
fr1.set_ang_vel(fr0, u1 * fr0.z)

# origin of inertial frame
Nor = me.Point('N_or')

# joints
jt0 = Nor.locatenew('jt_0', q3 * frN.x + q4 * frN.y)
jt1 = jt0.locatenew('jt_1', l0 * fr0.x)
jt2 = jt1.locatenew('jt_2', l1 * fr1.x)

# mass centers
mc0 = jt0.locatenew('mc_0', l0 / 2 * fr0.x)
mc1 = jt1.locatenew('mc_1', l1 / 2 * fr1.x)

# constrained point velocities
Nor.set_vel(frN, 0)
jt0.set_vel(frN, u3 * frN.x + u4 * frN.y)

# unconstrained point velocities
jt1.v2pt_theory(jt0, frN, fr0)
jt2.v2pt_theory(jt1, frN, fr1)

mc0.v2pt_theory(jt0, frN, fr0)
mc1.v2pt_theory(jt1, frN, fr1)

# inertia dyads
in0 = (me.inertia(fr0, 0, 0, Izz0), mc0)
in1 = (me.inertia(fr1, 0, 0, Izz1), mc1)

# bodies
bd0 = me.RigidBody('bd_0', mc0, fr0, m0, in0)
bd1 = me.RigidBody('bd_1', mc1, fr1, m1, in1)

# kinematic differential equations
kd = [q0.diff(t) - u0,
      q1.diff(t) - u1,
      q3.diff(t) - u3,
      q4.diff(t) - u4]

# forces
fl = [(mc0, -m0 * g * frN.z),
      (mc1, -m1 * g * frN.z)]

# fl = [(mc0, -m0 * g * frN.y),
#       (mc1, -m1 * g * frN.y)]

# bodies
bl = [bd0, bd1]

# q_ind and u_ind
qind = [q0, q3, q4]
uind = [u0, u3, u4]

qdep = [q1]
udep = [u1]

qcons = [q1 - f1]
ucons = [u1 - f1d]

# equations of motion using Kane's method
kane = me.KanesMethod(frame=frN,
                      q_ind=qind,
                      u_ind=uind,
                      kd_eqs=kd,
                      q_dependent=qdep,
                      configuration_constraints=qcons,
                      u_dependent=udep,
                      velocity_constraints=ucons)

(fr, frstar) = kane.kanes_equations(fl, bl)
kanezero = fr + frstar
```

```python
n = 3

# group variables
x, y, theta = me.dynamicsymbols('x y theta')  # generalized coordinates
xd, yd, thetad = me.dynamicsymbols('x y theta', 1)
us, vs, omega = me.dynamicsymbols('u v omega')  # generalized speeds

# shape or configuration variables
q = me.dynamicsymbols('q:' + str(n - 1))
u = me.dynamicsymbols('u:' + str(n - 1))

# constraints
cq = me.dynamicsymbols('c:' + str(n - 1))  # holonomic constraint
cu = me.dynamicsymbols('c:' + str(n - 1), 1)  # non-holonomic constraint
cs = me.dynamicsymbols('c:' + str(n - 1), 2)  # used in matrix substitution later

# forces and torques
F = me.dynamicsymbols('F:' + str(n + 1))  # n + 1 joints to apply forces/torques
T = me.dynamicsymbols('T:' + str(n + 1))

# variables for bodies
g, t = sym.symbols('g t')
m = sym.symbols('m:' + str(n))
l = sym.symbols('l:' + str(n))
Izz = sym.symbols('I_zz_:' + str(n))

# frames
frames = [me.ReferenceFrame('fr_N')]
for i in range(n):
    name = 'fr_' + str(i)
    if i == 0:
        frame = frames[i].orientnew(name, 'Axis', [theta, frames[i].z])
        frame.set_ang_vel(frames[i], omega * frames[i].z)
    else:
        frame = frames[i].orientnew(name, 'Axis', [q[i - 1], frames[i].z])
        frame.set_ang_vel(frames[i], u[i - 1] * frames[i].z)
    frames.append(frame)

# origin of inertial frame
joints = [me.Point('N_or')]
joints[0].set_vel(frames[0], 0)

# joints
for i in range(n + 1):
    name = 'jt_' + str(i)
    if i == 0:
        joint = joints[i].locatenew(name, x * frames[0].x + y * frames[0].y)
        joint.set_vel(frames[0], us * frames[0].x + vs * frames[0].y)
    else:
        joint = joints[i].locatenew(name, l[i - 1] * frames[i].x)
        joint.v2pt_theory(joints[i], frames[0], frames[i])
    joints.append(joint)

# mass centers
mc = []
for i in range(n):
    name = 'mc_' + str(i)
    mcent = joints[i + 1].locatenew(name, l[i] / 2 * frames[i + 1].x)
    mcent.v2pt_theory(joints[i + 1], frames[0], frames[i + 1])
    mc.append(mcent)

# inertia dyads
dyads = []
for i in range(n):
    dyad = (me.inertia(frames[i + 1], 0, 0, Izz[i]), mc[i])
    dyads.append(dyad)

# bodies
bodies = []
for i in range(n):
    name = 'bd_' + str(i)
    body = me.RigidBody(name, mc[i], frames[i + 1], m[i], dyads[i])
    bodies.append(body)

# kinematic differential equations
kd = [xd - us, yd - vs, thetad - omega]
for i in range(n - 1):
    kd.append(q[i].diff(t) - u[i])
    
# independent coordinates and speeds
qind = [x, y, theta]
uind = [us, vs, omega]

# dependent coordinates and speeds
qdep = q
udep = u

# constraints (they must equal zero)
# Note: qcons (holonomic constraints) only matter for linearization
qcons = []
ucons = []
for i in range(n - 1):
    qcons.append(q[i] - cq[i])
    ucons.append(u[i] - cu[i])

# gravitational forces
#TODO: add aerodynamic forces
forces = []
for i in range(n):
    forces.append((mc[i], -m[i] * g * frames[0].z))

torques = []
for i in range(n):
    torques.append((frames[i + 1], (T[i] - T[i + 1]) * frames[0].z))

fl = forces #+ torques

# setup dictionaries of values to return
dynamic = dict(qind=qind, uind=uind,
               qdep=qdep, udep=udep,
               qcons=qcons, ucons=ucons,
               q=qind + qdep, u=uind + udep,
               kd=kd, T=T, F=F,
               cq=cq, cu=cu, cs=cs)
symbol = dict(m=m, l=l, Izz=Izz, g=g, t=t)
setup = dict(frames=frames, joints=joints, mc=mc, dyads=dyads)
fbd = dict(fl=fl, bodies=bodies)
```

```python
# equations of motion using Kane's method
kane = me.KanesMethod(frame=setup['frames'][0],
                      q_ind=dynamic['qind'],
                      u_ind=dynamic['uind'],
                      kd_eqs=dynamic['kd'],
                      q_dependent=dynamic['qdep'],
                      configuration_constraints=dynamic['qcons'],
                      u_dependent=dynamic['udep'],
                      velocity_constraints=dynamic['ucons'])

# kane = me.KanesMethod(frame=frames[0],
#                       q_ind=qind,
#                       u_ind=uind,
#                       kd_eqs=kd,
#                       q_dependent=qdep,
#                       configuration_constraints=qcons,
#                       u_dependent=udep,
#                       velocity_constraints=ucons)

(fr, frstar) = kane.kanes_equations(fl, bodies)
kanezero = fr + frstar
mass = kane.mass_matrix_full
forcing = kane.forcing_full

eom = dict(kane=kane, fr=fr, frstar=frstar, mass=mass, forcing=forcing, kanezero=kanezero)
```

```python
#F_func, M_func, ffunc, mfunc, dummy_dict = nchain.callable_matrices(dynamic, eom, consts)

# dummy symbols used for dynamamic variables
dyn_symbols = dynamic['cs'] + dynamic['q'] + dynamic['u']
dummy_symbols = [sym.Dummy() for i in dyn_symbols]
dummy_dict = dict(zip(dyn_symbols, dummy_symbols))

# substitute dummy symbols into mass matrix and forcing vector
kd_dict = eom['kane'].kindiffdict()  # solved kinematical differential equations
F = eom['forcing'].subs(kd_dict).subs(dummy_dict)  # substitute into the forcing vector
M = eom['mass'].subs(kd_dict).subs(dummy_dict)  # substitute into the mass matrix 

# callable matrices
F_func = sym.lambdify(consts['names'] + dummy_symbols, F)
M_func = sym.lambdify(consts['names'] + dummy_symbols, M)

# partially evaluate matrices to run faster (maybe 15% or so)
ffunc = functools.partial(F_func, *consts['vals'])
mfunc = functools.partial(M_func, *consts['vals'])
```

```python
def generate_trajectory(thi, thf, ti, tf, n=50, end=False):
    
    tfnorm = tf - ti
    
    a0 = thi
    a1 = 0
    a2 = 3 / tfnorm**2 * (thf - thi)
    a3 = -2 / tfnorm**3 * (thf - thi)

    
    t = np.linspace(0, tfnorm, n, endpoint=end)
    th = a0 + a1 * t + a2 * t**2 + a3 * t**3
    dth = a1 + 2 * a2 * t + 3 * a3 * t**2
    ddth = 2 * a2 + 6 * a3 * t
    
    return np.c_[t + ti, th, dth, ddth]


def quintic_trajectory(thi, thf, ti, tf, dthi=0, dthf=0, ddthi=0, ddthf=0, n=50, end=False):
    """Quinitic trajectory from Craig, p. 209. This should
    be smoother.
    """
    
    tfn = tf - ti
    
    a0 = thi
    a1 = dthi
    a2 = ddthi / 2
    a3 = (20 * thf - 20 * thi - (8 * dthf + 12 * dthi) * tfn - (3 * ddthi - ddthf) * tfn**2) / (2 * tfn**3)
    a4 = (30 * thi - 30 * thf + (14 * dthf + 16 * dthi) * tfn + (3 * ddthi - 2 * ddthf) * tfn**2) / (2 * tfn**4)
    a5 = (12 * thf - 12 * thi - (6 * dthf + 6 * dthi) * tfn - (ddthi - ddthf) * tfn**2) / (2 * tfn**5)
    
    t = np.linspace(0, tfn, n, endpoint=end)
    #t = np.linspace(ti, tf, n, endpoint=end)
    th = a0 + (a1 * t) + (a2 * t**2) + (a3 * t**3) + (a4 * t**4) + (a5 * t**5)
    dth = a1 + (2 * a2 * t) + (3 * a3 * t**2) + (4 * a4 * t**3) + (5 * a5 * t**4)
    ddth = 2 * a2 + (6 * a3 * t) + (12 * a4 * t**2) + (20 * a5 * t**3)
    
    return np.c_[ti + t, th, dth, ddth]


def path_gen(traj_gen, angles, times, npts):
    """Generate a quintic path.
    """
    
    assert len(angles) == len(times)
    
    nparts = len(angles)
    traj = []

    for i in range(nparts - 1):
        if i == nparts - 1:
            end = True
        else:
            end = False

        part = traj_gen(angles[i], angles[i+1], times[i], times[i+1], n=npts, end=end)
        traj.append(part)

    tr = np.array(traj)
    tr = tr.reshape(-1, 4)

    return tr


def path_repeat(path, ntimes):
    """Repeat a path ntimes.
    """
    
    print('this is not finished yet.')
    return
    
    tf, ti = path[-1, 0], path[0, 0]
    dt = path[1, 0] - ti
    
    path = np.tile(path.T, ntimes).T
    path[:, 0] = np.linspace(ti, ntimes * tf, len(path), endpoint=False)
    
    return path


_test = generate_trajectory(15, 75, 1, 3, end=True)
_test = quintic_trajectory(15, 75, 1, 3, end=True)
tm, th, dth, ddth = _test.T
```

```python
def organize_data(y, n, snake_dict):
    """Organize the data so it can be animated with ellipses.
    """

    ms = snake_dict['m']
    ls = snake_dict['l']
    npts = y.shape[0]
    angs = y[:, 2:n + 2]
    rads = angs.cumsum(axis=1)
    degs = np.rad2deg(rads)

    jtx, jty = y[:, :2].T
    jtx = jtx.reshape(-1, 1)
    jty = jty.reshape(-1, 1)

    mcx, mcy = [], []
    for i in range(n):
        dx = ls[i] * np.cos(rads[:, i])
        dy = ls[i] * np.sin(rads[:, i])

        mcx.append((jtx[:, i] + dx / 2).flatten())
        mcy.append((jty[:, i] + dy / 2).flatten())

        x = (jtx[:, i] + dx).reshape(-1, 1)
        y = (jty[:, i] + dy).reshape(-1, 1)

        jtx = np.hstack((jtx, x))
        jty = np.hstack((jty, y))

    mcx = np.array(mcx).T
    mcy = np.array(mcy).T

    # center of mass
    mms = np.array(ms)
    comx = (mcx * mms).sum(axis=1) / mms.sum()
    comy = (mcy * mms).sum(axis=1) / mms.sum()
    
    if False:
        jtx = (jtx.T - np.abs(comx)).T
        jty = (jty.T - np.abs(comy)).T
        mcx = (mcx.T - np.abs(comx)).T
        mcy = (mcy.T - np.abs(comy)).T
        comx = comx - np.abs(comx)
        comy = comy - np.abs(comy)

    return rads, degs, jtx, jty, mcx, mcy, comx, comy



def n_movie_maker(n, snake_dict, ts, y, limx=(-1, 1), limy=(-1, 1), hst=50, interval=100):

    rads, degs, jtx, jty, mcx, mcy, comx, comy = organize_data(y, n, snake_dict)

    ms = snake_dict['m']
    ls = snake_dict['l']
    ws = snake_dict['w'] * 2.5
    
    ii = 0
    st = 0

    fig, ax = plt.subplots(1, 1, sharex=True, sharey=True, figsize=(9, 9))
    ppl.plot(ax, [], [])
    ax.axis('image')
    ax.set_xlim(limx)
    ax.set_ylim(limy)

    fig.set_facecolor('w')
    #fig.tight_layout()

    time_fmt = '{0:2.3f} sec'
    time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes, fontsize='large')

    ellipse_params = dict(alpha=.5, color='green')
    patches = []
    for i in range(n):
        e = Ellipse(xy=(mcx[ii, i], mcy[ii, i]),
                    width=ws[i], height=1.05 * ls[i],
                    angle=degs[ii, i] - 90, **ellipse_params)
        ax.add_patch(e)
        patches.append(e)

    head, = ax.plot(jtx[ii, 0], jty[ii, 0], 'go', alpha=.6, markersize=7)
    jlines, mclines = [], []
    for i in range(n + 1):
        jline, = ax.plot(jtx[st:ii, i], jty[st:ii, i], 'b--', alpha=.5)
        jlines.append(jline)
    for i in range(n):
        mcline, = ax.plot(mcx[st:ii, i], mcy[st:ii, i], 'r', alpha=.5)
        mclines.append(mcline)

    coml, = ax.plot(comx[st:ii], comy[st:ii], 'k-', lw=2, alpha=.9)

    def init():
        ii = 0
        head.set_data(jtx[ii, 0], jty[ii, 0])
        for i in range(n):
            patches[i].set_alpha(.3)
        time_text.set_text('')
        return [head, time_text] + patches

    def animate(ii):

        st = ii - hst
        if st < 0:
            st = 0

        head.set_data(jtx[ii, 0], jty[ii, 0])
        for i in range(n):
            patches[i].set_visible(True)
            patches[i].set_alpha(.5)
            patches[i].center = (mcx[ii, i], mcy[ii, i])
            patches[i].angle = degs[ii, i] - 90
            jlines[i].set_data(jtx[st:ii, i], jty[st:ii, i])
            mclines[i].set_data(mcx[st:ii, i], mcy[st:ii, i])
        for i in range(n + 1):
            jlines[i].set_data(jtx[st:ii, i], jty[st:ii, i])

        time_text.set_text(time_fmt.format(ts[ii]))

        coml.set_data(comx[st:ii], comy[st:ii])

        return patches + [head, coml, time_text] + jlines + mclines

    anim = animation.FuncAnimation(fig, animate, frames=len(ts),
                interval=interval, blit=False, repeat=False, init_func=init)

    return anim
```
